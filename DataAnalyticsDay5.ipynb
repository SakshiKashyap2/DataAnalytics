```python
# Write a program to print a plot 4 function f(x)=x^2 and f(x)=x^3 and f(x)=2x+4 and f(x)=4x^2+5x+4
# x is a list of values taken from the user and passed as paramter to the function myPlot
# set the limit of the axis such that x axis will be (min of x value in the list)-4 and y will be (max of x value)+4
# Give apprpriate legends, title, x label and y label and save the figure set the marker size as 10.5 and the linewidth as 2.5

import matplotlib.pyplot as plt

def myPlot(xList):
  l1=[]
  l2=[]
  l3=[]
  l4=[]
  for i in xList:
    func1=(i*i)
    func2=(i*i*i)
    func3=2*i+4
    func4=4*(i*i)+5*(i)+4

    l1.append(func1)
    l2.append(func2)
    l3.append(func3)
    l4.append(func4)
  plt.plot(xList,l1, 'ro--', markersize=10.5, linewidth=2.5, label='x vs x^2')
  plt.plot(xList,l2, 'b*--', markersize=10.5, linewidth=2.5, label='x vs x^3')
  plt.plot(xList,l3, 'gx--', markersize=10.5, linewidth=2.5, label='x vs 2x+4')
  plt.plot(xList,l4, 'c+--', markersize=10.5, linewidth=2.5, label='x vs 4x^2+5x+4')
  plt.title('Multiple function of x')
  plt.xlabel('x co-ordinate-------->')
  plt.ylabel('f(x)co-ordinate-------->')
  # plt.axis([0, min(xList)-4, 0, max(xList)+4])

  plt.legend()

  plt.show()

def main():
  myPlot([0,1,2,3,4,5])


main()
```

```python
#4) Using python, consider n random numbers in the range 0.345 to 0.567. 
#Write the numbers in a file.
#Next open the file in read mode and display its content, also display the average value. 
#Pass that value to a function for calculating area of a square.

import random


def writeFile(n):
  file=open('RandomNumber.txt', 'w')
  for i in range(0,n):
    randNum=random.uniform(0.345, 0.567)
    #print(round(randNum, 3))
    file.write(str(round(randNum, 3)))
    file.write('\n')
  file.close()

def displayAverage():
  file=open('RandomNumber.txt', 'r')
  print('File content:')
  sum=0
  count=0
  for each in file:
    print(each)
    sum=sum+float(each)
    count=count+1
  return round(sum/count, 3)
  file.close()

def square(n):
  return n*n

def main():
  n=eval(input('Enter number of random numbers:'))
  writeFile(n)
  avg=displayAverage()
  print('Average:', avg)
  print('Area of the square:', round(square(avg), 3))
main()
```

```python
#7) Consider a company which needs to display the number of units sold per Month for each product using multiline plots. 
#(i.e., Separate Plotline for each product). 
#The products are faceCreamSalesData, faceWashSalesData, toothPasteSalesData, bathingsoapSalesData,
#shampooSalesData, moisturizerSalesData. 
#Provide appropriate legend and place the legend in upper left using loc property. 
#Use plt.xticks to provide labels for the months in the x axis , and plt.yticks to set the values of y axis.


import matplotlib.pyplot as plt

def plotLine(data, label, pointerType):
  
  month=[1,2,3,4,5]
  plt.title('My Title')
  plt.plot(month,data, pointerType, markersize=10.5, linewidth=2.5, label=label)
  plt.xticks(month, ['Jan','Feb','Mar','Apr','May'], rotation ='vertical')
  plt.legend(loc='upper left')
  
def main():
  fig=plt.figure()
  faceCreamSalesData=[1,2,8,4,5]
  faceWashSalesData=[2,8,6,4,5]
  toothPasteSalesData=[6,9,3,1,5]
  bathingsoapSalesData=[8,6,2,1,5]
  shampooSalesData=[8,6,2,1,5]
  moisturizerSalesData=[8,6,5,8,6]

  plotLine(faceCreamSalesData, 'x vs FaceCreamSalesData', 'r+--')
  plotLine(faceWashSalesData, 'x vs FaceWashSalesData', 'b*--')
  plotLine(toothPasteSalesData, 'x vs ToothPasteSalesData', 'gx--')
  plotLine(bathingsoapSalesData, 'x vs BathingsoapSalesData', 'co--')
  plotLine(shampooSalesData, 'x vs ShampooSalesData', 'kv--')
  plotLine(moisturizerSalesData, 'x vs MoisturizerSalesData', 'm^--')
  plt.savefig('Question7.png')


main()
```

```python
#8) Modify the above program to display the number of units sold per month for the product
#faceCream product and display it with bar plot, using plt.bar(). 
#Use appropriate labelings.

import matplotlib.pyplot as plt

def plotLine(data, label, pointerType):
  month=[1,2,3,4,5]
  plt.title('My Title')
  plt.plot(month,data, pointerType, markersize=10.5, linewidth=2.5, label=label)
  plt.xticks(month, ['Jan','Feb','Mar','Apr','May'], rotation ='vertical')
  plt.legend(loc='upper left')

def plotBar(data):
  month=[1,2,3,4,5]
  plt.bar(month, data)

def main():
  fig=plt.figure()
  faceCreamSalesData=[1,2,8,4,5]
  faceWashSalesData=[2,8,6,4,5]
  toothPasteSalesData=[6,9,3,1,5]
  bathingsoapSalesData=[8,6,2,1,5]
  shampooSalesData=[8,6,2,1,5]
  moisturizerSalesData=[8,6,5,8,6]

  plotBar(faceCreamSalesData)
  plotLine(faceWashSalesData, 'x vs FaceWashSalesData', 'b*--')
  plotLine(toothPasteSalesData, 'x vs ToothPasteSalesData', 'gx--')
  plotLine(bathingsoapSalesData, 'x vs BathingsoapSalesData', 'co--')
  plotLine(shampooSalesData, 'x vs ShampooSalesData', 'kv--')
  plotLine(moisturizerSalesData, 'x vs MoisturizerSalesData', 'm^--')
  plt.savefig('Question8.png')

main()
```
