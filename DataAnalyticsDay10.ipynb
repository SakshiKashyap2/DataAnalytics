Classification Algorithm

Steps:
1. Preprocessing of the data
2. Creating training and test data set
3. Instantiating the classifier to be used
4. Trainging the classifier with the prepared traing data set
5. Making prediction of the classes and object to which object / sample will be launched
6.  Evaluation the performence of the result obtained by the classifier
7.  Tracking the performance matrix

Algorithms we have here, 
1. K - nearest neighbour (KNN)
2. Support vector machine (SVM)
3. Decision Tree/ Random Forest
4. Naive baise

```python
from sklearn.metrics._plot.confusion_matrix import confusion_matrix
import pandas as pd
import sklearn as sk
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.metrics import accuracy_score
from sklearn.tree import export_text

data=pd.read_csv('/content/iris.csv')
x=data.iloc[:, :-1] #x is original dataset
y=data['species'] #y is class level dataset
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)

#print('x train: ')
#print(x_train)
#print('y train: ')
#print(y_train)

DT_model=tree.DecisionTreeClassifier()
SVC_model=SVC()
KNN_model=KNeighborsClassifier(n_neighbors=5) #the parameter is number of neighbours, if not given it will take default value

#Model Training
SVC_model.fit(x_train, y_train)
KNN_model.fit(x_train, y_train)
DT_model.fit(x_train, y_train)

#Prediction
SVC_predict=SVC_model.predict(x_test)
KNN_predict=KNN_model.predict(x_test)
DT_predict=DT_model.predict(x_test)

#Tessting
#Test result (y_test)
print('Y Test: ') #Printing y_test and others for manual comparison
print(y_test)
print('SVC:')
print(SVC_predict)
print('KNN')
print(KNN_predict)
print('DT')
print(DT_predict)

#Printing score
print('SVC prediction score:', accuracy_score(SVC_predict, y_test))
print('KNN prediction score:', accuracy_score(KNN_predict, y_test))
print('DT prediction score:', accuracy_score(DT_predict, y_test))

#Confusion Matrix:
print('\nSVC predict confusion matrix: \n', sk.metrics.confusion_matrix(SVC_predict, y_test))
print('\nKNN predict confusion matrix: \n', sk.metrics.confusion_matrix(KNN_predict, y_test))
print('\nDT predict confusion matrix: \n', sk.metrics.confusion_matrix(DT_predict, y_test))

```

```python
from sklearn.tree import export_text

iris=load_iris()
x,y = iris.data, iris.target
clf=tree.DecisionTreeClassifier(criterion='entropy', max_depth=2) #max_depth means till given level we will get the tree
clf=clf.fit(x, y)

fig=plt.figure()
tree.plot_tree(clf)
fig.savefig('testDT.png')

r=export_text(clf, feature_names=iris['feature_names'])
print(r)
```
